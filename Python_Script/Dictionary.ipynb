## Mengcopy Dictionary
dalam mengcopy Dictionary harus menggunakan sebuah method ".copy()" karena jika tidak, saat kita mengcopy suatu data dictionary ke dalam variabel baru
saat variabel yang lama di update maka variabel baru juga akan ikut terupdate jika tidak menggunakan method tersebut sama hal nya dengan list.

```python
my_copy_dict = my_data.copy() # seperti ini tidak akan berubah jika variabel yang kita copy di update
my_copy_dict
```

```python
my_copy = my_data # ini akan berubah jika variabel yang di copy di update
my_copy
```
