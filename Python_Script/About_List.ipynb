## Menambah Elemen atau Method di List

1. append()
   sebuah elemen menambahkan data baru dengan posisi index baru yaitu terakhir

```python
negara.append("malaysia")
negara
```

2. insert(index:new_data)
   insert memasukan data dan manaruh di index yang kita inginkan

```python
negara.insert(1,"singapura")
negara
```

## Sorting List

```python
nomor = [1,5,2,4,3,9,6,8,7,10]
nomor
```

```python
nomor.sort() # mengurutkan angka terkecil - terbesar
nomor
```

```python
nomor.sort(reverse=True)  # mengurtkan angka terbesar - terkecil
nomor
```

## Mengupdate Elemen di List

```python
nomor[-1] = 10
nomor
```
