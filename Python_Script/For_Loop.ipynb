# For Loop

```python
for <variable> in <list>:
    <code>
```

```python
negara = ["cina","korea","amerika","canada"]
negara
```

```python
for ngr in negara:
    print(ngr)
```

```python
negara
```

```python
for i, ngr in enumerate(negara): #menyediakan indeks bersama dengan nilai (elemen) saat mengiterasi melalui sebuah iterable (seperti list, tuple, atau string).
    print(i,ngr)
```

```python
# Looping dalam Dcitionary
my_data = {'nama':'Haerul','umur':20}
my_data
```

```python
for key,value in my_data.items(): # untuk mengambil item harus menggunakan method .items()
    print(f"{key} : {value}")
```
